## 1-2-1. 単純なオートエンコーダー

### 画像を６４次元の潜在ベクトルに圧縮（次元削減）する Encoder
### +
### 潜在空間から元画像を再構成する Decoder
### = 
### オートエンコーダー

### モチベーション：
### * 入力を圧縮して復元する際に、なるべく元の特徴を損なわずに復元したい（再構成誤差を最小化するようにEncoderを学習したい）
### * そのように学習されたEncoderを別のタスクに利用したり、
### * 圧縮して得られた潜在表現そのものを利用したりしたい

```python
# 単純なEncoder, Decoder

latent_dim = 64 

class Autoencoder(Model):
    
  def __init__(self, latent_dim):
    
    super(Autoencoder, self).__init__()
    
    self.latent_dim = latent_dim # 潜在ベクトルの次元数
    
    self.encoder = tf.keras.Sequential([
                                                          layers.Flatten(), # 平滑化（28 × 28 → 784）
                                                          layers.Dense(latent_dim, activation='relu'), # 次元削減（784 → 64）
                                                         ])
    
    self.decoder = tf.keras.Sequential([
                                                          layers.Dense(784, activation='sigmoid'), # 復元（64 → 784）
                                                          layers.Reshape((28, 28)) # 整型（28 × 28 → 784）
                                                         ])

    
  def call(self, x): # encoder → decoder の処理をつなぐ
    
    encoded = self.encoder(x) 
    decoded = self.decoder(encoded) # encoderの出力を渡す
    
    return decoded
  

# インスタンス化    
autoencoder = Autoencoder(latent_dim) 
```

```python
# コンパイル 
autoencoder.compile(optimizer='adam', loss=losses.MeanSquaredError())
```

### Encoder はMNISTデータセットを784次元から64次元に削減することを学習
### Decoder はMNISTデータセットを64次元から784次元に復元することを学習

```python
autoencoder.fit(x_train, x_train,
                        epochs=10,
                        shuffle=True,
                        validation_data=(x_test, x_test)
                        )
```

```python
# トレーニングされたオートエンコーダーを、テストデータで試す

encoded_imgs = autoencoder.encoder(x_test).numpy()
decoded_imgs = autoencoder.decoder(encoded_imgs).numpy()
```

## 1-2-2. デノイジング・オートエンコーダー

```python
(x_train, _), (x_test, _) = fashion_mnist.load_data()
```

```python
x_train = x_train.astype('float32') / 255.
x_test = x_test.astype('float32') / 255.

x_train = x_train[..., tf.newaxis]
x_test = x_test[..., tf.newaxis]

print(x_train.shape)
```

```python
# 画像にランダムノイズを追加する

noise_factor = 0.2
x_train_noisy = x_train + noise_factor * tf.random.normal(shape=x_train.shape) 
x_test_noisy = x_test + noise_factor * tf.random.normal(shape=x_test.shape) 

x_train_noisy = tf.clip_by_value(x_train_noisy, clip_value_min=0., clip_value_max=1.)
x_test_noisy = tf.clip_by_value(x_test_noisy, clip_value_min=0., clip_value_max=1.)
```

```python
# ノイズが付与された画像の描画

n = 10
plt.figure(figsize=(20, 2))

for i in range(n):
    ax = plt.subplot(1, n, i + 1)
    plt.title("original + noise")
    plt.imshow(tf.squeeze(x_test_noisy[i]))
    plt.gray()
    
plt.show()
```

```python
# 畳み込みオートエンコーダーを定義

class Denoise(Model):
    
  def __init__(self):
    
    super(Denoise, self).__init__()
    self.encoder = tf.keras.Sequential([
                                                          layers.Input(shape=(28, 28, 1)),
                                                          layers.Conv2D(16, (3, 3), activation='relu', padding='same', strides=2),
                                                          layers.Conv2D(8, (3, 3), activation='relu', padding='same', strides=2)
                                                         ])

    self.decoder = tf.keras.Sequential([
                                                          layers.Conv2DTranspose(8, kernel_size=3, strides=2, activation='relu', padding='same'),
                                                          layers.Conv2DTranspose(16, kernel_size=3, strides=2, activation='relu', padding='same'),
                                                          layers.Conv2D(1, kernel_size=(3, 3), activation='sigmoid', padding='same')
                                                         ])
    
    
  def call(self, x):
    
    encoded = self.encoder(x)
    decoded = self.decoder(encoded)
    return decoded


autoencoder = Denoise()
```

```python
autoencoder.compile(optimizer='adam', loss=losses.MeanSquaredError())
```

```python
autoencoder.fit(x_train_noisy, x_train,
                        epochs=10,
                        shuffle=True,
                        validation_data=(x_test_noisy, x_test)
                        )
```

```python
# Encoderでは画像サイズが 28×28 から 7×7 にダウンサンプリングされている

autoencoder.encoder.summary()
```

```python
# Decoderでは画像サイズが 7×7 から 28×28 にアップサンプリングされている

autoencoder.decoder.summary()
```

## 1-2-3. 異常検出

### 心電図の異常なリズム（０）と正常なリズム（１）をセグメントしたい
### 正常なリズム（１）のみで学習した検出器は、異常なリズム（０）を入力すると大きい値の再構成誤差を出力する

```python
# データセットをロード

dataframe = pd.read_csv('http://storage.googleapis.com/download.tensorflow.org/data/ecg.csv', header=None)
raw_data = dataframe.values
dataframe.head()
```

```python
# raw_data の末尾にラベル情報が含まれている（4998, ）
labels = raw_data[:, -1]

# raw_data の末尾以外は、心電図の情報（4998, 141）
data = raw_data[:, 0:-1]


train_data, test_data, train_labels, test_labels = train_test_split(
                                                                                                data, labels, test_size=0.2, random_state=21
                                                                                                )
```

```python
# 正規化

min_val = tf.reduce_min(train_data)
max_val = tf.reduce_max(train_data)

train_data = (train_data - min_val) / (max_val - min_val)
test_data = (test_data - min_val) / (max_val - min_val)

train_data = tf.cast(train_data, tf.float32)
test_data = tf.cast(test_data, tf.float32)
```

```python
# 正常なリズム（１）のみで学習し、異常なリズムから分離する

train_labels = train_labels.astype(bool)
test_labels = test_labels.astype(bool)

normal_train_data = train_data[train_labels]
normal_test_data = test_data[test_labels]

anomalous_train_data = train_data[~train_labels]
anomalous_test_data = test_data[~test_labels]
```

```python
# 正常なリズム（１）のプロット

plt.grid()
plt.plot(np.arange(140), normal_train_data[0])
plt.title("A Normal ECG")
plt.show()
```

```python
# 異常なリズム（０）のプロット

plt.grid()
plt.plot(np.arange(140), anomalous_train_data[0])
plt.title("An Anomalous ECG")
plt.show()
```

```python
class AnomalyDetector(Model):
    
  def __init__(self):
    
    super(AnomalyDetector, self).__init__()
    
    self.encoder = tf.keras.Sequential([
                                                          layers.Dense(32, activation="relu"),
                                                          layers.Dense(16, activation="relu"),
                                                          layers.Dense(8, activation="relu")
                                                         ])

    self.decoder = tf.keras.Sequential([
                                                          layers.Dense(16, activation="relu"),
                                                          layers.Dense(32, activation="relu"),
                                                          layers.Dense(140, activation="sigmoid")
                                                         ])

    
  def call(self, x):
    
    encoded = self.encoder(x)
    decoded = self.decoder(encoded)
    return decoded


autoencoder = AnomalyDetector()
```

```python
autoencoder.compile(optimizer='adam', loss='mae')
```

```python
history = autoencoder.fit(normal_train_data, normal_train_data,  # 自身を目的変数として学習
                                       epochs=20, 
                                       batch_size=512,
                                       validation_data=(test_data, test_data),
                                       shuffle=True
                                       )
```

```python
# そもそもLossは小さい状態から始まる

plt.plot(history.history["loss"], label="Training Loss")
plt.plot(history.history["val_loss"], label="Validation Loss")
plt.legend()
```

```python
# 正常なリズムと、再構成後の正常なリズムとの比較
# 波形は大体同じ

encoded_imgs = autoencoder.encoder(normal_test_data).numpy()
decoded_imgs = autoencoder.decoder(encoded_imgs).numpy()

plt.plot(normal_test_data[0], 'b') # 通常の心電図
plt.plot(decoded_imgs[0], 'r') # 再構成後の心電図
plt.fill_between(np.arange(140), decoded_imgs[0], normal_test_data[0], color='lightcoral')
plt.legend(labels=["Input", "Reconstruction", "Error"])
plt.show()
```

```python
# 異常なリズムと、再構成後の異常なリズムとの比較
# 再構成後のリズムは正常なリズムと似た波形になる為、異常なリズムとの誤差が大きくなる

encoded_imgs = autoencoder.encoder(anomalous_test_data).numpy()
decoded_imgs = autoencoder.decoder(encoded_imgs).numpy()

plt.plot(anomalous_test_data[0], 'b')
plt.plot(decoded_imgs[0], 'r')
plt.fill_between(np.arange(140), decoded_imgs[0], anomalous_test_data[0], color='lightcoral')
plt.legend(labels=["Input", "Reconstruction", "Error"])
plt.show()
```

```python
# 再構成誤差のプロット

reconstructions = autoencoder.predict(normal_train_data)
train_loss = tf.keras.losses.mae(reconstructions, normal_train_data)

plt.hist(train_loss, bins=50)
plt.xlabel("Train loss")
plt.ylabel("No of examples")
plt.show()
```

```python
threshold = np.mean(train_loss) + np.std(train_loss)
print("Threshold: ", threshold)
```

```python
reconstructions = autoencoder.predict(anomalous_test_data)
test_loss = tf.keras.losses.mae(reconstructions, anomalous_test_data)

plt.hist(test_loss, bins=50)
plt.xlabel("Test loss")
plt.ylabel("No of examples")
plt.show()
```

```python
def predict(model, data, threshold):
    
  reconstructions = model(data)
  loss = tf.keras.losses.mae(reconstructions, data)
  return tf.math.less(loss, threshold)


def print_stats(predictions, labels):
    
  print("Accuracy = {}".format(accuracy_score(labels, preds)))
  print("Precision = {}".format(precision_score(labels, preds)))
  print("Recall = {}".format(recall_score(labels, preds)))
```

```python
"""
TensorFlowで実装したニューラルネットワークを使いIrisデータセットを3クラス分類する
"""

from sklearn.preprocessing import OneHotEncoder

df = pd.read_csv("Iris.csv")


# データフレームから条件抽出
y = df["Species"]
X = df.loc[:, ["SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalWidthCm"]]

# NumPy 配列に変換
X = np.array(X)
y = np.array(y)

# ラベルを数値に変換
y[y == "Iris-setosa"] = 0
y[y == "Iris-versicolor"] = 1
y[y == "Iris-virginica"] = 2

y = y.astype(np.int64)[:, np.newaxis]

enc = OneHotEncoder(handle_unknown='ignore', sparse=False)

y = enc.fit_transform(y)


# trainとtestに分割
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# さらにtrainとvalに分割
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=0)
```

```python
# ハイパーパラメータの設定
learning_rate = 0.001
batch_size = 10
num_epochs = 100
n_hidden1 = 50
n_hidden2 = 100
n_input = X_train.shape[1]
n_samples = X_train.shape[0]
n_classes = 3
```

### 7-1. Iris（2値分類）

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.autograd import Variable

# データセットの読み込み
df = pd.read_csv("Iris.csv")

# データフレームから条件抽出
df = df[(df["Species"] == "Iris-versicolor") | (df["Species"] == "Iris-virginica")]
y = df["Species"]
X = df.loc[:, ["SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalWidthCm"]]

# NumPy 配列に変換
X = np.array(X)
y = np.array(y)

# ラベルを数値に変換
y[y == "Iris-versicolor"] = 0
y[y == "Iris-virginica"] = 1
y = y.astype(np.int64)[:, np.newaxis]
```

```python
# torch に変換
X = torch.from_numpy(X.astype(np.float32)).clone()
y = torch.from_numpy(y.astype(np.float32)).clone()
```

```python
# trainとtestに分割
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# さらにtrainとvalに分割
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=0)
```

```python
# ハイパーパラメータの設定

learning_rate = 0.001
batch_size = 10
num_epochs = 100
n_hidden1 = 50
n_hidden2 = 100
n_input = X_train.shape[1]
n_samples = X_train.shape[0]
n_classes = 1
```

```python
# TensorDataset と DataLoader を使う
from torch.utils.data import TensorDataset
from torch.utils.data import DataLoader

train_ds = TensorDataset(X_train, y_train)
train_dl = DataLoader(train_ds, batch_size=batch_size, shuffle=True)

valid_ds = TensorDataset(X_val, y_val)
valid_dl = DataLoader(valid_ds, batch_size=batch_size*2)
```

```python
model = nn.Sequential(
                                   nn.Linear(n_input, n_hidden1),
                                   nn.ReLU(),
                                   nn.Linear(n_hidden1, n_hidden2),
                                   nn.ReLU(),
                                   nn.Linear(n_hidden2, n_classes),
                                   nn.Sigmoid()
                                   )
```

```python
#勾配法
optimizer = optim.SGD(model.parameters(), lr=learning_rate)

#誤差関数
loss_func = nn.BCELoss()
```

```python
# 学習

# 学習ループ
for epoch in range(num_epochs):
    
    model.train()
    
    for xb, yb in train_dl:
        
        # 順伝播
        out = model(xb)

        # ロスの計算
        loss = loss_func(out, yb)

        # 勾配の計算
        loss.backward()

        # パラメータの更新
        optimizer.step()

        # 勾配の初期化
        optimizer.zero_grad()
    
    model.eval()
    with torch.no_grad():
        val_loss  = sum(loss_func(model(xb), yb) for xb, yb in valid_dl)
        
    print(epoch, val_loss / len(valid_dl))
```

```python
# 学習V2

def loss_batch(model, loss_func, xb, yb, opt=None):
    loss = loss_func(model(xb), yb)
    
    if opt is not None:
        loss.backward()
        optimizer.step()
        optimizer.zero_grad()
        
    return loss.item(), len(xb)
```

```python
def fit(epochs, model, loss_func, opt, train_dl, valid_dl):
    
    loss_hist = []
    
    for epoch in range(epochs):
        
        model.train()
        
        for xb, yb in train_dl:
            loss_batch(model, loss_func, xb, yb, opt)

        model.eval()
        
        with torch.no_grad():
            losses, nums = zip(*[loss_batch(model, loss_func, xb, yb) for xb, yb in valid_dl])
            
        val_loss = np.sum(np.multiply(losses, nums)) / np.sum(nums)
        loss_hist.append(val_loss)
            
        print(epoch, val_loss)
        
    return loss_hist
```

```python
def get_data(train_ds, valid_ds, bs):
    
    return (
                DataLoader(train_ds, batch_size=bs, shuffle=True),
                DataLoader(valid_ds, batch_size=bs * 2),
                )

train_dl, valid_dl = get_data(train_ds, valid_ds, batch_size)
```

```python
%%time

hist = fit(num_epochs, model, loss_func, optimizer, train_dl, valid_dl)
```

  * シンプルで使いやすい、とっつきやすい

  * TensorFlow は構造が複雑で機能の全体像が掴みづらい


* Pytorch
   * torch.tensor を使い配列の操作・変換が必要
   * 複雑且つ可読性が低い、研究者向け
   * TensorFlow とできることはあまり変わらないが、モデルの低レベルのカスタムがしやすく柔軟性に富む
