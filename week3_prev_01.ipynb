# 【問題1】データの取得

```python
from sklearn.datasets import load_iris
import pandas as pd

data = load_iris()
X = pd.DataFrame(data.data, columns=["sepal_length", "sepal_width", "petal_length", "petal_width"])
X
```

```python
y = pd.DataFrame(data.target, columns=["Species"])
y
```

# 【問題2】データの結合

```python
df = pd.concat([X, y], axis=1)
df
```

# 【問題6】図の作成

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# "Species"0 = setosa

df_setosa =  df.loc[df["Species"] == 0]

plt.figure(figsize=(10, 10))
plt.title("setosa")
labels=df_setosa.iloc[:, 0:4].columns
plt.pie(df_setosa.iloc[:, 0:4].sum(), labels=labels, counterclock=False, startangle=90, autopct="%1.1f%%")

plt.show()
```

* 全ての組み合わせの散布図を一度に表示させる、散布図行列を作る(１つ)

```python
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
plt.subplots_adjust(wspace=0.25, hspace=0.4)

sns.scatterplot(ax=axes[0, 0], x="sepal_length", y="sepal_width", data=df, hue="Species", palette="Set2")
axes[0, 0].set_title("sepal_length & sepal_width")

sns.scatterplot(ax=axes[0, 1], x="sepal_length", y="petal_length", data=df, hue="Species", palette="Set2")
axes[0, 1].set_title("sepal_length & petal_length")

sns.scatterplot(ax=axes[0, 2], x="sepal_length", y="petal_width", data=df, hue="Species", palette="Set2")
axes[0, 2].set_title("sepal_length & petal_width")

sns.scatterplot(ax=axes[1, 0], x="sepal_width", y="petal_length", data=df, hue="Species", palette="Set2")
axes[1, 0].set_title("sepal_width & petal_length")

sns.scatterplot(ax=axes[1, 1], x="sepal_width", y="petal_width", data=df, hue="Species", palette="Set2")
axes[1, 1].set_title("sepal_width & petal_width")

sns.scatterplot(ax=axes[1, 2], x="petal_length", y="petal_width", data=df, hue="Species", palette="Set2")
axes[1, 2].set_title("petal_length & petal_width")

plt.show()
```

# 【問題8】可視化したグラフや表の結果の説明

* データを探索した時には、結果を述べる（+考察する）ことが大切です。ここまでで作成した表や図から読み取れることや気づきを説明してください。例えば、散布図行列を見ながら、3種類のアヤメにはどのような違いがあるかを説明してください。


* 箱ひげ図・バイオリン図から
  + 基本的には virginica > versicolor > setosa の順で、各特徴量の値が大きい。
  + setosa は萼片の幅 (sepal_width) のみ、他の品種より値が大きい。
  + setosa は花びらのサイズ (petal_length, petal_width) の分布の幅は狭い。
  
* 散布図から
  + versicolor と virginica の分布は似ていて、基本的に各特徴量同士に正の相関を持つ。　
    - 萼片の長さと花びらの長さ(sepal_length & petal_length)、花びらの長さと花びらの幅(petal_length & petal_width)の相関は強い。
    - 萼片の長さと萼片の幅(sepal_length & sepal_width)の相関は弱い。
  + setosa は基本的に各特徴量同士にはあまり相関がない。
    - 萼片の長さと萼片の幅(sepal_length & sepal_width) には正の相関がある。
 
* ヒートマップから
  + 全体として、萼片の幅(sepal_width) は他の特徴量との相関が比較的弱めである。
