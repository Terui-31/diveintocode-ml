# 【問題5】用語の調査

* 尖度

>分布が正規分布からどれだけ尖っているかを表す統計量で、山の尖り度と裾の広がり度を示します。サンプルサイズを$n$、各データ$x_i (i\colon 1, 2, \cdots, n)$の平均値を$\overline{x}$、標準偏差を$s$とすると尖度は次の式から求められます。
$$
\frac{n(n+1)}{(n-1)(n-2)(n-3)} \sum_{i = 1}^n \frac{(x_i - \overline{x})^4}{s^4} - \frac{3(n-1)^2}{(n-2)(n-3)} 
$$
正規分布より尖った分布（データが平均付近に集中し、分布の裾が重い）のときには正の値を、正規分布より扁平な分布（データが平均付近から散らばり、分布の裾が軽い）のときには負の値をとります。正規分布の場合には0になります。

* 歪度

>分布が正規分布からどれだけ歪んでいるかを表す統計量で、左右対称性を示す指標のことです。サンプルサイズを$n$, 各データ$x_i(i: 1, 2, \cdots, n)$の平均値を$\bar{x}$、標準偏差をsとすると歪度は次式で求められます。
$$
\frac{n}{(n-1)(n-2)} \sum_{i=1}^{n} \Bigl( \frac{x_i - \bar{x}}{s} \Bigr)^3
$$
「右裾が長い」もしくは「右に歪んだ」もしくは「左に偏った」分布のときには正の値を、「左裾が長い」もしくは「左に歪んだ」もしくは「右に偏った」分布のときには負の値をとります。左右対称の分布（例えば正規分布）の場合には0になります。

https://bellcurve.jp/statistics/course/17950.html

# 【問題6】分布の確認

* seabornのsns.distplot()を使い目的変数の分布を表示し、「尖度」と「歪度」も算出する。

```python
# 目的変数の分布

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
sns.distplot(train.iloc[:, -1])

plt.show()
```

```python
# 歪度(Skewness)と尖度(Kurtosis)

y = "SalePrice"

print("Skewness :", train[y].skew())
print("Kurtosis :", train[y].kurt())
```

* 目的変数に対して対数変換を行う。

```python
import numpy as np

# +1は発散防止
log_y = np.log(train[y]+1)
log_y
```

* 対数変換したものに対して分布を表示し、「尖度」と「歪度」も算出する。

```python
# 分布を表示

plt.figure(figsize=(12, 8))
sns.distplot(log_y)

plt.show()
```

```python
# 尖度と歪度を算出

print("Skewness :", log_y.skew())
print("Kurtosis :", log_y.kurt())
```

対数変換の前後での変化を「正規分布」という言葉を含めて説明してください。このように加工することで予測時の精度を上げることができます。

→ 対数変換を用いることで値を正規化し、過度に非対称に歪んだ分布の尖度や歪度を抑えて正規分布に従う変数に変換した。


# 【問題7】相関係数の確認

* 相関係数行列のヒートマップを作り、相関係数の高い関係を見つける。

```python
plt.figure(figsize=(20, 16))
sns.heatmap(train.corr())

plt.show()
```

>* 車庫関係の特徴量など、似たジャンルの特徴量は相関が高めなことが確認できる。
* 竣工年と車庫の竣工年など、「常識」的に考えて同じになりやすい特徴量もある。
* 値が正規化されていないので、単位が似た特徴量も相関が高くなりやすい。

* 目的変数との相関の高い特徴量10個を選び、これらだけの相関係数行列ヒートマップを作る。

```python
# 目的変数との相関係数の一覧（降順）

train.corr().iloc[-1].sort_values()[::-1]
```

```python
# そのtop10（自身は除く）

train.corr().iloc[-1].sort_values()[::-1][1:11]
```

```python
# top10のインデックスを取得

top10_index = train.corr().iloc[-1].sort_values()[::-1][1:11].index
top10_index
```

```python
# top10のインデックスを使って特徴量を抽出

top10_train = train.loc[:, top10_index]
top10_train
```

```python
# 相関係数ヒートマップ作成

plt.figure(figsize=(12, 8))
sns.heatmap(top10_train.corr(), annot=True)

plt.show()
```

* 選んだ10個の特徴量が何か表しているかをKaggleのDataDescriptionの記述を参考に日本語でまとめる。

>* OverallQual…家の全体的な材質と仕上がりの評価（10段階評価）
* GrLivArea…地上階の延床居住面積（単位は平方フィート）
* GarageCars…車庫の収容可能台数
* GarageArea…車庫のサイズ（単位は平方フィート）
* TotalBsmtSF…地下室の面積の合計（単位は平方フィート）
* 1stFlrSF…１階の面積（単位は平方フィート）
* FullBath…フルバス（浴槽・シャワー・トイレ・洗面台のセット）の数
* TotRmsAbvGrd…バスルームを除いた部屋数の合計数
* YearBuilt…竣工年
* YearRemodAdd…増築・改築年（増改築無しなら竣工年と同一）

* 選んだ10個の特徴量の中でお互いの相関係数が高い組み合わせを3つ探し出す。

>* GarageCars と GarageArea
* TotalBsmtSF と 1stFlrSF
* GrLivArea と TotRmsAbvGrd
  + これらの相関係数が高い特徴量同士は、一つの特徴量として集約するか、片方を落とすと良いかと思われる。
